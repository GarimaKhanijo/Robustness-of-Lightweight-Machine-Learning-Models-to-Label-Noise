```python
# cell 2

import numpy as np
import pandas as pd

from noise import add_symmetric_noise, add_asymmetric_noise
```

```python
# cell 3

columns = [
"age","workclass","fnlwgt","education","education-num",
"marital-status","occupation","relationship","race","sex",
"capital-gain","capital-loss","hours-per-week","native-country","income"
]

df_train = pd.read_csv("../data/adult/adult.data",
                       names=columns,
                       na_values=" ?",
                       skipinitialspace=True)

df_test = pd.read_csv("../data/adult/adult.test",
                      names=columns,
                      na_values=" ?",
                      skiprows=1,
                      skipinitialspace=True)

df = pd.concat([df_train, df_test], ignore_index=True)

df = df.dropna()

# clean label formatting
df["income"] = df["income"].str.replace(".", "", regex=False)

# encode target variable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(df["income"])

print("Original class distribution:")
print("Class 0:", np.sum(y == 0))
print("Class 1:", np.sum(y == 1))
```

```python
# cell 4

y_sym = add_symmetric_noise(y, noise_rate=0.2)

print("After 20% Symmetric Noise:")
print("Class 0:", np.sum(y_sym == 0))
print("Class 1:", np.sum(y_sym == 1))
```

```python
# cell 5

y_asym = add_asymmetric_noise(y, noise_rate=0.2)

print("After 20% Asymmetric Noise (1 -> 0):")

print("Class 0:", np.sum(y_asym == 0))
print("Class 1:", np.sum(y_asym == 1))
```
